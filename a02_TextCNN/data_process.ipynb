```python
import pandas as pd
```

```python
questions = pd.read_csv('../../question_train_word.csv')
```

```python
questions_topics = questions.topics.apply(lambda s: s.split(','))
questions_titles = questions.titles.astype('U').apply(lambda s: s.split(','))
questions_descriptions = questions.description.astype('U').apply(lambda s: s.split(','))
```

```python
print(list(map(lambda x: len(x), [questions_descriptions, questions_titles, questions_topics])))
```

```python
# w1 dw2 __label__id
d_len = [len(d) for d in questions_descriptions]
```

```python
from collections import Counter
c = Counter(d_len)
```

```python
c.most_common(10)
```

```python
from tqdm import tqdm

result = []
for i in tqdm(range(len(questions_titles))):
    result.append(' '.join(questions_titles[i]) + ' '
                  + ' '.join(questions_descriptions[i][-50:])
                  + '__label__' + ' '.join([str(t) for t in questions_topics[i]]))
```

```python
result[0]
```

```python
with open('train_w_dw.txt', 'w') as f:
    for i in tqdm(range(len(result))):
        f.write(result[i] + '\n')
```
